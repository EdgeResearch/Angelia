```python
global_sentiment = []
global_sentiment_original = []
for item in news.iterrows():
    sum = (item[1]['sentiment_transformers'] + item[1]['sentiment_textblob'] + item[1]['sentiment_sentiment'] + item[1]['sentiment_natural'])
    global_sentiment_original.append(sum)
    if sum > 0:
        global_sentiment.append(1)
    elif sum < 0:
        global_sentiment.append(-1)
    else:
        global_sentiment.append(0)
    
news['global_sentiment'] = global_sentiment
news['global_sentiment_original'] = global_sentiment_original
```

```python
def negativity_score_evaluator(evidence_score, neutrality_score, contraddictory_score, w_same, w_opposite):
    """if evidence_score - contraddictory_score == 0:
        return 1/evidence_score"""
    
    if evidence_score-neutrality_score == 0:
        return False
  
    #if w_same + w_opposite != 0:
        #evidence_score = evidence_score*(w_same + w_opposite)
        
    #contraddictory_score = contraddictory_score * w_opposite
    same_polarity_score = (evidence_score - contraddictory_score - neutrality_score)
    #same_polarity_score = same_polarity_score * w_same
    #SC = evidence_score - contraddictory_score - neutrality_score
    PSC = (same_polarity_score / (evidence_score - neutrality_score)) * 100
    #PSC = ((contraddictory_score + neutrality_score) / (evidence_score)) * 100

    #PNEU = (neutrality_score / evidence_score) * 100
    #PFC = (contraddictory_score / evidence_score) * 100
    FCP = (100 - (PSC))
    
    
    return (1 - ((contraddictory_score / evidence_score) * 100))

```

```python
def w_evaluator(current_evidence, item, type):
    counter = 0
    weight = 0
    same = []
    opposite = []
    for p in current_evidence.iterrows():
        if type == 1 and p[1]["global_sentiment"] == item[1]["global_sentiment"]:
            counter+=1
            weight += p[1]['global_sentiment_original']
            same.append(p[1]['global_sentiment_original'])
        elif type == -1 and (p[1]["global_sentiment"] != item[1]["global_sentiment"]) and p[1]["global_sentiment"] != 0:
            counter+=1
            opposite.append(p[1]['global_sentiment_original'])
            weight += p[1]['global_sentiment_original']
    
    if counter == 0:
        return 1
    
    if weight == 0:
        return 1
        
    return weight/counter  
        
    
```

```python
sentiment_to_use = "global_sentiment"
negativity_scores = []
for item in news.iterrows():
    current_evidence = evidence[evidence["news_i"] == item[1]["index"]]
    w_same = 0
    w_opposite = 0
        
    if len(current_evidence) >= 1:
        w_same = w_evaluator(current_evidence,item, 1)
        w_opposite = w_evaluator(current_evidence,item, -1)
        evidence_score = len(current_evidence)
        neutrality_score = len(current_evidence[current_evidence[sentiment_to_use] == 0])
        contraddictory_score = len(current_evidence[(current_evidence[sentiment_to_use] == item[1][sentiment_to_use]*(-1)) & (current_evidence[sentiment_to_use] != 0)])
        negativity_score = negativity_score_evaluator(evidence_score,neutrality_score,contraddictory_score, w_same, w_opposite)
        negativity_scores.append(negativity_score)
    else:
        negativity_scores.append(False)

news['negativity_score'] = negativity_scores
```

```python
min_boundary = 30
max_boundary = 60
```

```python
true_zone_f = 0
fake_zone_f = 0
boundary_zone_f = 0
num_fake_news = len(news[(news["label"] == "FAKE") & (news["negativity_score"] != False)])
for item in news[(news["label"] == "FAKE") & (news["negativity_score"] != False)].iterrows():
    if item[1]["negativity_score"] > max_boundary:
        fake_zone_f += 1
    elif item[1]["negativity_score"] < min_boundary:
        true_zone_f += 1
    else:
        boundary_zone_f += 1
        
print("FAKE NEWS")
print("FAKE: "+str(fake_zone_f/num_fake_news))
print("REAL: "+str(true_zone_f/num_fake_news))
print("BOUNDARY: "+str(boundary_zone_f/num_fake_news))
```

```python
true_zone_t = 0
fake_zone_t = 0
boundary_zone_t = 0
num_real_news = len(news[(news["label"] == "REAL") & (news["negativity_score"] != False)])
for item in news[(news["label"] == "REAL") & (news["negativity_score"] != False)].iterrows():
    if item[1]["negativity_score"] > max_boundary:
        fake_zone_t += 1
    elif item[1]["negativity_score"] < min_boundary:
        true_zone_t += 1
    else:
        boundary_zone_t += 1
        
print("REAL NEWS")
print("FAKE: "+str(fake_zone_t/num_real_news))
print("REAL: "+str(true_zone_t/num_real_news))
print("BOUNDARY: "+str(boundary_zone_t/num_real_news))
```
